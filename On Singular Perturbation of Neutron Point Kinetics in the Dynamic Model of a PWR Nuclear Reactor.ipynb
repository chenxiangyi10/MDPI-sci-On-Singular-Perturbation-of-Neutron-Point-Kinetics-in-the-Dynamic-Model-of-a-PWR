# Runge-Kutta-Fehlberg Method

```python
import numpy as np


def rkf(model, a, b, x0, tol, hmax, hmin, n):
    """Runge-Kutta-Fehlberg method to solve x' = f(x,t) with x(t[0]) = x0.
        Based on pseudocode presented in "Numerical Analysis", 6th Edition,
        by Burden and Faires, Brooks-Cole, 1997. And 9th Edition, by Burden and Faires, 2011.
    
    Acknowledgement: the method is developed by Xiangyi Chen <xxc90@psu.edu>
        based on the method written by:
        Jonathan Senning <jonathan.senning@gordon.edu>
        Gordon College

    USAGE:
        t, x = rkf(model, a, b, x0, tol, hmax, hmin, n)

    INPUT:
        model - a class contains the function equal to dx/dt = f(x,t) and algebraic equations
        a     - left-hand endpoint of interval (initial condition is here)
        b     - right-hand endpoint of interval
        x0    - initial x value: x0 = x(a)
        tol   - maximum value of local truncation error estimate
        hmax  - maximum step size
        hmin  - minimum step size
        n     - number of differential equations

    OUTPUT:
        t     - NumPy array of independent variable values
        x     - NumPy array of corresponding solution function values
        r     - 

    NOTES:
        This function implements 4th-5th order Runge-Kutta-Fehlberg Method
        to solve the initial value problem

           dx
           -- = f(x,t),     x(a) = x0
           dt

        on the interval [a,b].
    """

    # Coefficients used to compute the independent variable argument of f

    a2 = 2.500000000000000e-01  # 1/4
    a3 = 3.750000000000000e-01  # 3/8
    a4 = 9.230769230769231e-01  # 12/13
    a5 = 1.000000000000000e+00  # 1
    a6 = 5.000000000000000e-01  # 1/2

    # Coefficients used to compute the dependent variable argument of f

    b21 = 2.500000000000000e-01  # 1/4
    b31 = 9.375000000000000e-02  # 3/32
    b32 = 2.812500000000000e-01  # 9/32
    b41 = 8.793809740555303e-01  # 1932/2197
    b42 = -3.277196176604461e+00  # -7200/2197
    b43 = 3.320892125625853e+00  # 7296/2197
    b51 = 2.032407407407407e+00  # 439/216
    b52 = -8.000000000000000e+00  # -8
    b53 = 7.173489278752436e+00  # 3680/513
    b54 = -2.058966861598441e-01  # -845/4104
    b61 = -2.962962962962963e-01  # -8/27
    b62 = 2.000000000000000e+00  # 2
    b63 = -1.381676413255361e+00  # -3544/2565
    b64 = 4.529727095516569e-01  # 1859/4104
    b65 = -2.750000000000000e-01  # -11/40

    # Coefficients used to compute local truncation error estimate.  These
    # come from subtracting a 4th order RK estimate from a 5th order RK
    # estimate.

    r1 = 2.777777777777778e-03  # 1/360
    r3 = -2.994152046783626e-02  # -128/4275
    r4 = -2.919989367357789e-02  # -2197/75240
    r5 = 2.000000000000000e-02  # 1/50
    r6 = 3.636363636363636e-02  # 2/55

    # Coefficients used to compute 4th order RK estimate

    c1 = 1.157407407407407e-01  # 25/216
    c3 = 5.489278752436647e-01  # 1408/2565
    c4 = 5.353313840155945e-01  # 2197/4104
    c5 = -2.000000000000000e-01  # -1/5

    # Set t and x according to initial condition and assume that h starts
    # with a value that is as large as possible.

    t = a
    x = x0
    h = hmax

    # Initialize arrays that will be returned

    T = np.array([t])
    X = np.array([x])
    R = np.array([])  # the estimate of truncation error

    x = x0[0:n]
    while t < b:
        x = x[0:n]

        # Adjust step size when we get to last interval

        if t + h > b:
            h = b - t

        # Compute values needed to compute truncation error estimate and
        # the 4th order RK estimate.
        # unsupported operand type(s) for *: 'float' and 'NoneType'
        k1 = h * model.differential_eqns(x, t)
        k2 = h * model.differential_eqns(x + b21 * k1, t + a2 * h)
        k3 = h * model.differential_eqns(x + b31 * k1 + b32 * k2, t + a3 * h)
        k4 = h * \
            model.differential_eqns(
                x + b41 * k1 + b42 * k2 + b43 * k3, t + a4 * h)
        k5 = h * \
            model.differential_eqns(
                x + b51 * k1 + b52 * k2 + b53 * k3 + b54 * k4, t + a5 * h)
        k6 = h * model.differential_eqns(x + b61 * k1 + b62 *
                                         k2 + b63 * k3 + b64 * k4 + b65 * k5, t + a6 * h)

        # Compute the estimate of the local truncation error.  If it's small
        # enough then we accept this step and save the 4th order estimate.

        r = abs(r1 * k1 + r3 * k3 + r4 * k4 + r5 * k5 + r6 * k6) / h
        if len(np.shape(r)) > 0:
            r = max(r)
            # print(r)
        if r <= tol:
            t = t + h
            x = x + c1 * k1 + c3 * k3 + c4 * k4 + c5 * k5
            x = model.append_states_of_algebraic_eqns(x, t)
            T = np.append(T, t)
            X = np.append(X, [x], 0)
            R = np.append(R, r)

        # Now compute next step size, and make sure that it is not too big or
        # too small.

        h = h * min(max(0.84 * (tol / r) ** 0.25, 0.1), 4.0)

        if h > hmax:
            h = hmax
        elif h < hmin:
            print("Error: stepsize should be smaller than %e." % hmin)
            break

    # endwhile

    return (T, X, R)
```

# Defining reactor core model

## Define reactor core model - without using singular perturbation

```python
class Reactor_core_model:
    def __init__(self):
        self.ff = 0.98  # fraction of energy deposited in the fuel
        self.Omega = 6.53  # MW/degree-C
        self.M = 92.8  # MW/degree-C
        self.mu_f = 26.3  # MW.s/degree-C
        self.mu_c = 70.5  # MW.s/degree-C
        self.Gr = 0.01    # this control rod reactivity [not used in this code]
        self.alpha_c = 0.00001  # reactivity/degree-C
        self.alpha_f = -0.00005  # reactivity/degree-C
        self.P0a = 2500  # MW
        # self.Pc0 = 2450  # MW
        self.Te = 290  # degree-C, inlet temperature
        # degree-C, initial fuel and structure temperature, 678.66 degree-C
        self.Tf0 = self.Te + self.P0a / \
            (2 * self.M) + self.ff * self.P0a / self.Omega
        self.Tl0 = self.Te + self.P0a / self.M  # 316.9 degree-C
        # degree-C, initial coolant temperature, 304.8
        self.Tc0 = (self.Tl0 + self.Te) / 2
        self.rho_r0 = 0  # control rod initial reactivity?
        self.nr0 = 1  # initial value of relative neutron density
        self.lambds = np.array(
            [0.0124, 0.0305, 0.1110, 0.3010, 1.1400, 3.0100])
        #self.Lambd = 0.0000179
        self.Lambd = 0.0001
        self.Beta = 0.006502
        self.betas = np.array(
            [0.000215, 0.001424, 0.001274, 0.002568, 0.000748, 0.000273])
        self.rho = 0  # reactivity
        self.rho_r = 0  # control rod reactivity
        self.nr = 1  # neutron density (relative)
        # state variables initial values
        # [nr, cr1, cr2, cr3, cr4, cr5, cr6, Tf, Tl, control_rod_reactivity, Reactivity]
        self.x0 = np.array([1, 1, 1, 1, 1, 1, 1, self.Tf0, self.Tl0, 0, 0])

    def differential_eqns(self, x, t):
        # x0 = nr, x1 = cr, x2 = Tf, x3 = Tl
        x0 = x[0]  # nr
        x1 = x[1:7]  # cr
        x2 = x[7]  # Tf
        x3 = x[8]  # Tl
        # control rod reactivity at time t
        rho_r = self.control_rod_reactivity(t)
        rho = rho_r + self.alpha_f * \
            (x2 - self.Tf0) + self.alpha_c * \
            ((x3+self.Te) / 2 - self.Tc0)  # overall reactivity
        dnrdt = (rho - self.Beta) / self.Lambd * \
            x0 + sum(self.betas / self.Lambd * x1)
        dcrdt = self.lambds * x0 - self.lambds * x1
        dTfdt = (self.ff * self.P0a * x0 - self.Omega *
                 (x2 - (x3 + self.Te) / 2)) / self.mu_f
        dTldt = ((1 - self.ff) * self.P0a * x0 + self.Omega *
                 (x2 - (x3 + self.Te)/2) - self.M * (x3 - self.Te)) / self.mu_c
        return np.concatenate((dnrdt, dcrdt, dTfdt, dTldt), axis=None)

    def append_states_of_algebraic_eqns(self, x, t):
        x0 = x[0]  # nr
        x1 = x[1:7]  # cr
        x2 = x[7]  # Tf
        x3 = x[8]  # Tl
        rho_r = self.control_rod_reactivity(t)
        rho = rho_r + self.alpha_f * \
            (x2 - self.Tf0) + self.alpha_c * \
            ((x3+self.Te) / 2 - self.Tc0)  # overall reactivity
        return np.concatenate((x, rho_r, rho), axis=None)
```

## Defining reactor core model - using singular perturbation

```python
class Reactor_core_model_perturbed:
    def __init__(self):
        self.ff = 0.98  # fraction of energy deposited in the fuel
        self.Omega = 6.53  # MW/degree-C
        self.M = 92.8  # MW/degree-C
        self.mu_f = 26.3  # MW.s/degree-C
        self.mu_c = 70.5  # MW.s/degree-C
        self.Gr = 0.01
        self.alpha_c = 0.00001  # reactivity/degree-C
        self.alpha_f = -0.00005  # reactivity/degree-C
        self.P0a = 2500  # MW
        self.Pc0 = 2450  # MW
        self.Te = 290  # degree-C, inlet temperature
        # degree-C, initial fuel and structure temperature, 678.66 degree-C
        self.Tf0 = self.Te + self.P0a / \
            (2 * self.M) + self.ff * self.P0a / self.Omega
        self.Tl0 = self.Te + self.P0a / self.M  # 316.9 degree-C
        # degree-C, initial coolant temperature, 304.8
        self.Tc0 = (self.Tl0 + self.Te) / 2
        self.rho_r0 = 0  # control rod initial reactivity?
        self.nr0 = 1  # initial value of relative neutron density
        self.lambds = np.array(
            [0.0124, 0.0305, 0.1110, 0.3010, 1.1400, 3.0100])
        #self.Lambd = 0.0000179
        self.Lambd = 0.0001
        self.Beta = 0.006502
        self.betas = np.array(
            [0.000215, 0.001424, 0.001274, 0.002568, 0.000748, 0.000273])
        self.rho = 0  # reactivity
        self.rho_r = 0  # control rod reactivity
        self.nr = 1  # neutron density (relative)
        # state variables initial values
        # [cr1, cr2, cr3, cr4, cr5, cr6, Tf, Tl, nr, control_rod_reactivity, Reactivity]
        self.x0 = np.array([1, 1, 1, 1, 1, 1, self.Tf0, self.Tl0, 1, 0, 0])

    def differential_eqns(self, x, t):
        # x0 = nr, x1 = cr, x2 = Tf, x3 = Tl
        # x0 = x[0] # nr
        x1 = x[0:6]  # cr
        x2 = x[6]  # Tf
        x3 = x[7]  # Tl
        # control rod reactivity at time t
        rho_r = self.control_rod_reactivity(t)
        rho = rho_r + self.alpha_f * \
            (x2 - self.Tf0) + self.alpha_c * \
            ((x3+self.Te) / 2 - self.Tc0)  # overall reactivity
        nr = -sum(self.betas / (rho - self.Beta) * x1)
        dcrdt = self.lambds * nr - self.lambds * x1
        dTfdt = (self.ff * self.P0a * nr - self.Omega *
                 (x2 - (x3 + self.Te) / 2)) / self.mu_f
        dTldt = ((1 - self.ff) * self.P0a * nr + self.Omega *
                 (x2 - (x3 + self.Te)/2) - self.M * (x3 - self.Te)) / self.mu_c
        return np.concatenate((dcrdt, dTfdt, dTldt), axis=None)

    def append_states_of_algebraic_eqns(self, x, t):
        # x0 = x[0] # nr
        x1 = x[0:6]  # cr
        x2 = x[6]  # Tf
        x3 = x[7]  # Tl
        rho_r = self.control_rod_reactivity(t)
        rho = rho_r + self.alpha_f * \
            (x2 - self.Tf0) + self.alpha_c * \
            ((x3+self.Te) / 2 - self.Tc0)  # overall reactivity
        nr = -sum(self.betas / (rho - self.Beta) * x1)
        return np.concatenate((x, nr, rho_r, rho), axis=None)
```

# Without singular perturbation simulation

## One step reactivity insertion

```python
reactor_core_o = Reactor_core_model()


def control_rod_reactivity(t):
    return 2e-4


reactor_core_o.control_rod_reactivity = control_rod_reactivity
```

```python
if __name__ == "__main__":
    from pylab import *

    # record time duration
    currentDT1 = datetime.datetime.now()
    print("Starting time: ", str(currentDT1))

    starting_time, end_time = (0.0, 0.1)
    x0 = np.array(
        [1, 1, 1, 1, 1, 1, 1, reactor_core_o.Tf0, reactor_core_o.Tl0, 0, 0])
    tol = 1e-6

    t_rkf_o, x_rkf_o, r_rkf_o = rkf(reactor_core_o, starting_time,
                                    end_time, x0, tol, hmax=10.0, hmin=0.00001, n=9)  # unequally spaced t
    currentDT2 = datetime.datetime.now()
    print("Steps used: ", t_rkf_o.size)
    print("End time: ", str(currentDT2))
    print("Duration: ", (currentDT2 - currentDT1).total_seconds(), 'seconds')
```

### Plot the results

```python
plt.rcParams["figure.figsize"] = [6*2, 5*5]
subplot(4, 2, 1)
plot(t_rkf_o, x_rkf_o[:, 0])
xlabel('$t$')
ylabel('$x$')
title('Relative Power')

subplot(4, 2, 2)
plot(t_rkf_o, x_rkf_o[:, 1:7])
xlabel('$t$')
ylabel('$x$')
title('Relative Precursors')

subplot(4, 2, 3)
plot(t_rkf_o, x_rkf_o[:, 7] / reactor_core_o.Tf0)
xlabel('$t$')
ylabel('$x$')
title('Relative Fuel temperature')

subplot(4, 2, 4)
plot(t_rkf_o, x_rkf_o[:, 8] / reactor_core_o.Tl0)
xlabel('$t$')
ylabel('$x$')
title('Relative Outlet temperature')


subplot(4, 2, 5)
plot(t_rkf_o, x_rkf_o[:, 9])
xlabel('$t$')
ylabel('$x$')
title('Control Rod reactivity')


subplot(4, 2, 6)
plot(t_rkf_o, x_rkf_o[:, 10])
xlabel('$t$')
ylabel('$x$')
title('Overall reactivity')


subplot(4, 2, 7)
plot(t_rkf_o[1:], r_rkf_o)
xlabel('$t$')
ylabel('$|x - x^*|$')
title('Estimate of local truncation error (not the true error)')


subplot(4, 2, 8)
plot(t_rkf_o[1:], np.diff(t_rkf_o), 'b-o')
xlabel('$t$')
ylabel('$step size$')
title('Step size vs time')

ticklabel_format(useOffset=False)
show()
```

## Sinusoidal reactivity insertion

```python
if __name__ == "__main__":
    from pylab import *

    # record time duration
    currentDT1 = datetime.datetime.now()
    print("Starting time: ", str(currentDT1))

    starting_time, end_time = (0.0, 10.0)
    x0 = np.array(
        [1, 1, 1, 1, 1, 1, 1, reactor_core_s.Tf0, reactor_core_s.Tl0, 0, 0])
    tol = 1e-6

    t_rkf_s, x_rkf_s, r_rkf_s = rkf(reactor_core_s, starting_time,
                                    end_time, x0, tol, hmax=10.0, hmin=0.00001, n=9)  # unequally spaced t
    currentDT2 = datetime.datetime.now()
    print("Steps used: ", t_rkf_s.size)
    print("End time: ", str(currentDT2))
    print("Duration: ", (currentDT2 - currentDT1).total_seconds(), 'seconds')
```
